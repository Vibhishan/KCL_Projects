### Clinical research data store

Consider the following description of a system for storing and searching patient data for clinical studies:
1. A ```study``` is a collection of ```patients```
  1. A ```study``` has clinical owners and researchers who can access it
2. A ```patient``` is an individual participating in a study
  1. A ```patient``` has metadata relating to identifying them within the hospital system
  2. They may also have a few fields, such as age and gender that are relevant to the study
  3. A ```patient``` has ```sessions``` that correspond to a given examination or treatment
3. A ```session``` has metadata relating to time, date, type of session, and clinicians involved
  1. A ```session``` contains artefacts, such as datasets (imaging and non-imaging), assessors (data generated by processing scans) and reports
4. A ```dataset``` is a collection of files relating to a medical dataset, which could be a scan, genomic data, or some other kind of structured data
  1. ```dataset```s have metadata such as date, sequence (if MRI) parameters about the data, as well as thumbnails
  2. The actual data itself might be very large and stored elsewhere
5. A ```user``` is someone who can access this system
  1. Users cannot see all of the data, only the parts that they have permissions to view
  2. Some users cannot add / modify data
  3. Some users cannot delete data
  4. Administrators have permissions per project

**Question 1:** In your group, spend the next 20 minutes considering the following:
  1. Consider the entities in this system and the relationships between them
  2. Consider how you would structure this system in terms of modules/classes

### Use Case: Patients in multiple studies

A patient might be taking part in more than one study. When this is the case:
1. The system should be capable of recogising that the patient in one study is the same patient as the patient in another study
2. Not all sessions of a patient are relevant to all studies
  1. Each study should see only the sessions that are relevant to that study


**Question 2:** This raises an interesting issue. If, before, you had a study owning patients, you need to rethink the relationship
  * Try to consider several different mechanisms for resolving this issue

### Use Case: Everything in the system should have accessible history
A every entity in the system has a history that can be accessed
 * Changes don't overwrite records, but instead create a new version of them
 * Old versions must still be accessible
